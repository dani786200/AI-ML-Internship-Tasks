# Task 3: Heart Disease Prediction

## Objective
To build a machine learning model that predicts heart disease based on patient data.

## Dataset
UCI Heart Disease Dataset

## Steps
1. Data Loading
2. Data Cleaning
3. Exploratory Data Analysis
4. Model Training
5. Model Evaluation
6. Conclusion

```python
import pandas as pd

df = pd.read_csv("heart.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(x='target', data=df)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X = df.drop('target', axis=1)
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)//
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, pred))
print(confusion_matrix(y_test, pred))
```

## Conclusion
- A Logistic Regression model was used.
- The model achieved good accuracy.
- It can help predict heart disease risk based on health parameters.
