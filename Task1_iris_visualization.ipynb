## Objective
The objective of this task is to explore and visualize the Iris dataset to understand patterns, relationships, and distributions of data.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
df.head()
```

## Dataset Description
The Iris dataset contains 150 records of flowers with 4 features:
- Sepal Length
- Sepal Width
- Petal Length
- Petal Width

It also includes a target column called "species" which has 3 categories:
Setosa, Versicolor, Virginica.

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)
plt.title("Sepal vs Petal Length")
plt.show()
```

```python
df.hist(figsize=(8,6))
plt.show()
```

```python
sns.boxplot(data=df)
plt.title("Boxplot for Outliers")
plt.show()
```

## Conclusion

- Petal length and width clearly separate different species.
- Setosa species is easily distinguishable from others.
- Some minor outliers are present in sepal width.
- The dataset is well-structured and balanced.
